```python
import pandas as pd
import numpy as np
```

```python
movies['tags'] = movies['genres'] + movies['title']
movies.head()
```

```python
new_df = movies.drop(['genres', 'title'], axis = 1)
```

```python
new_df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=10000, stop_words='english')
```

```python
cv
```

```python
vec = cv.fit_transform(new_df['tags'].values.astype('U')).toarray()
```

```python
vec
```

```python
vec.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
sin = cosine_similarity(vec)
```

```python
sin
```

```python
dist = sorted((list(enumerate(sin[0]))), reverse=True,key=lambda vec:vec[1])
```

```python
for i in dist[0:5]:
  print(new_df.iloc[i[0]])
```

```python
def recommend(movies):
  index = new_df[new_df['title']==movies].index[0]
  distance = sin[index]
  for i in distance[0:5]:
    print(new_df.iloc[i[0]])
```
